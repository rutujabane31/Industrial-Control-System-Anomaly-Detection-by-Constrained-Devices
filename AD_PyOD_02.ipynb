```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
from pyod.models.ocsvm import OCSVM
from pyod.models.pca import PCA
from pyod.models.knn import KNN
from pyod.models.iforest import IForest
from pyod.utils.data import evaluate_print
import time


```

```python
# Load and preprocess the dataset for training
df_train = pd.read_csv('train1.csv')  # Replace 'your_train_dataset.csv' with the actual file name
train_data = df_train.drop(['Time', 'BLANK1', 'BLANK2', 'BLANK3'], axis=1)
train_data.dropna(axis=0, inplace=True)
X_train = train_data.values
```

```python
# Define the range of hyperparameters for each model
#ocsvm_params = {'nu': [0.1, 0.3, 0.5]}
ocsvm_params = {'kernel': ['linear', 'rbf'], 'gamma': ['scale', 'auto'], 'nu': [0.1, 0.2, 0.3]}
pca_params = {'n_components': [2, 4, 6]}
knn_params = {'n_neighbors': [3, 5, 7, 9 ]}
iforest_params = {'n_estimators': [50, 100, 150]}

# Perform hyperparameter tuning using grid search
models = {'OCSVM': (OCSVM(), ocsvm_params),
          'PCA': (PCA(), pca_params),
          'kNN': (KNN(), knn_params),
          'Isolation Forest': (IForest(), iforest_params)}

best_models = {}

for model_name, (model, param_grid) in models.items():
    grid_search = GridSearchCV(model, param_grid, scoring='accuracy', cv=5)
    grid_search.fit(X_train)
    best_model = grid_search.best_estimator_
    best_models[model_name] = best_model

# Assign the best models to variables
best_ocsvm = best_models['OCSVM']
best_pca = best_models['PCA']
best_knn = best_models['kNN']
best_iforest = best_models['Isolation Forest']

```

```python
# Load and preprocess the test datasets
df_test_01 = pd.read_csv('test1.csv')  # Replace 'your_test_dataset_01.csv' with the actual file name
df_test_02 = pd.read_csv('test2.csv')  # Replace 'your_test_dataset_02.csv' with the actual file name
test_data_01 = df_test_01.drop(['Time', 'BLANK1', 'BLANK2', 'BLANK3'], axis=1)
test_data_02 = df_test_02.drop(['Time', 'BLANK1', 'BLANK2', 'BLANK3'], axis=1)
test_data_01.dropna(axis=0, inplace=True)
test_data_02.dropna(axis=0, inplace=True)
X_test_01 = test_data_01.values
X_test_02 = test_data_02.values

```

```python
# Perform anomaly detection and calculate metrics for Test Dataset 1
# Initialize dictionaries to store the metrics and prediction times
metrics_01 = {'OCSVM': {}, 'PCA': {}, 'kNN': {}, 'Isolation Forest': {}}
prediction_times_01 = {'OCSVM': 0, 'PCA': 0, 'kNN': 0, 'Isolation Forest': 0}

for model_name, model in [('OCSVM', best_ocsvm), ('PCA', best_pca), ('kNN', best_knn), ('Isolation Forest', best_iforest)]:
    start_time = time.time()
    y_pred_01 = model.predict(X_test_01)
    end_time = time.time()
    prediction_time = end_time - start_time

    prediction_times_01[model_name] = prediction_time
    prediction_times_01[model_name] = round(prediction_time, 4)  # Round to 4 decimal places


    y_true_01 = np.ones_like(y_pred_01)
    tn, fp, fn, tp = confusion_matrix(y_true_01, y_pred_01, labels=[0, 1]).ravel()
    fpr = fp / (fp + tn)
    
    print(y_pred_01)
    print(y_true_01)
    metrics_01[model_name]['Accuracy'] = accuracy_score(y_true_01, y_pred_01)
    metrics_01[model_name]['Precision'] = precision_score(y_true_01, y_pred_01)
    metrics_01[model_name]['Recall'] = recall_score(y_true_01, y_pred_01)
    metrics_01[model_name]['F1 Score'] = f1_score(y_true_01, y_pred_01)
    metrics_01[model_name]['False Positive Rate'] = fpr

    metrics_01[model_name]['Accuracy'] = round(metrics_01[model_name]['Accuracy'], 6)  # Round to 6 decimal places
    metrics_01[model_name]['Precision'] = round(metrics_01[model_name]['Precision'], 6)  # Round to 6 decimal places
    metrics_01[model_name]['Recall'] = round(metrics_01[model_name]['Recall'], 6)  # Round to 6 decimal places
    metrics_01[model_name]['F1 Score'] = round(metrics_01[model_name]['F1 Score'], 6)  # Round to 6 decimal places
    metrics_01[model_name]['False Positive Rate'] = round(fpr, 6)  # Round to 6 decimal places

```

```python
# Perform anomaly detection and calculate metrics for Test Dataset 2
# Initialize dictionaries to store the metrics
metrics_02 = {'OCSVM': {}, 'PCA': {}, 'kNN': {}, 'Isolation Forest': {}}
prediction_times_02 = {'OCSVM': 0, 'PCA': 0, 'kNN': 0, 'Isolation Forest': 0}

# Calculate metrics for each model
for model_name, model in [('OCSVM', best_ocsvm), ('PCA', best_pca), ('kNN', best_knn), ('Isolation Forest', best_iforest)]:
    start_time = time.time()
    y_pred_02 = model.predict(X_test_02)
    end_time = time.time()
    prediction_time = end_time - start_time

    prediction_times_02[model_name] = prediction_time
    prediction_times_02[model_name] = round(prediction_time, 4)  # Round to 4 decimal places


    y_true_02 = np.ones_like(y_pred_02)
    tn, fp, fn, tp = confusion_matrix(y_true_02, y_pred_02, labels=[0, 1]).ravel()
    fpr = fp / (fp + tn)
    
    metrics_02[model_name]['Accuracy'] = accuracy_score(y_true_02, y_pred_02)
    metrics_02[model_name]['Precision'] = precision_score(y_true_02, y_pred_02)
    metrics_02[model_name]['Recall'] = recall_score(y_true_02, y_pred_02)
    metrics_02[model_name]['F1 Score'] = f1_score(y_true_02, y_pred_02)
    metrics_02[model_name]['False Positive Rate'] = fpr
    metrics_02[model_name]['Accuracy'] = round(metrics_02[model_name]['Accuracy'], 6)  # Round to 6 decimal places
    metrics_02[model_name]['Precision'] = round(metrics_02[model_name]['Precision'], 6)  # Round to 6 decimal places
    metrics_02[model_name]['Recall'] = round(metrics_02[model_name]['Recall'], 6)  # Round to 6 decimal places
    metrics_02[model_name]['F1 Score'] = round(metrics_02[model_name]['F1 Score'], 6)  # Round to 6 decimal places
    metrics_02[model_name]['False Positive Rate'] = round(fpr, 6)  # Round to 6 decimal places


```

```python
# Print the results
print("Results for Test Dataset 1:")
for model_name, metrics in metrics_01.items():
    print(model_name)
    for metric_name, value in metrics.items():
        print(metric_name + ":", value)
    print("Prediction Time:", prediction_times_01[model_name],"seconds")
    print()

print("Results for Test Dataset 2:")
for model_name, metrics in metrics_02.items():
    print(model_name)
    for metric_name, value in metrics.items():
        print(metric_name + ":", value)
    print("Prediction Time:", prediction_times_02[model_name],"seconds")
    print()
```
